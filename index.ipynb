## Your Task

Your task is to create a bar graph of the 25 most common words in Shakespeare's Macbeth.  


A common Python programming pattern to count objects, produce histograms, or update statistics is to make calls to a dictionary as you iterate through a list. For example, given a list of words, you can create a dictionary to store counts and then iterate through the list of words, checking how many times each word has appeared using your dictionary, and updating the dictionary count now that you've seen that word again. The `.get()` dictionary method is very useful in doing this. Read the docstring for the `.get()` method and use it along with the pseudocode below to create a bar graph of the 25 most common words from the transcript of Macbeth which has been loaded into the variable 'macbeth'. Be sure to include a title and appropriate labels for your graph.

To get the 25 *most common* words, you will have to sort your counts. If you are not super sure how to do this, checkout out the [Sorting HOW TO](https://docs.python.org/3/howto/sorting.html) Python documentation. Part of being a data scientist is figuring out how to do tasks that you may not have done before. Remember, in these situations, Google is your friend!

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#making everything lowercase and splitting the string into a list, then making a new variable called set_list
#that pulls all unique words from the lists and appends them.
mb_list = macbeth.lower()
mb_list = mb_list.split()

#Creating a blank dictionary
mb_dict = {}

#Official keys for the dictionary using the list of words from mb_list. Assigning values to keys based on the count
#for each word in mb_list
for word in mb_list:
    mb_dict[word] = mb_dict.get(word, 0) + 1

#Sorting the values from highest to lowest in the dictionary
top_values = sorted(mb_dict.values(), reverse=True)

#Placing the top 25 highest frequencies of keys into a list called top_25
top_25_values = top_values[:25]


#Placing the top 25 keys into a list called top_25_keys by iterating through the top_25_keys list and pushing it
#through the get_key() func to return each key and placing those keys into a list called top_25_keys
top_25_keys = []
def get_key(val): 
    for key, value in mb_dict.items(): 
         if val == value:
                return key

for value in top_25_values:
    key_of_val = get_key(value)
    top_25_keys.append(key_of_val)

#This is just to help me see the amount and number of keys - had an issue with these when I was trying to use the re library
print(f"Here is the length of keys: {len(top_25_keys)}\nHere is the length of values: {len(top_25_values)}")
print(f"List of keys: {top_25_keys}\nList of values: {top_25_values}")

#Creating a bar graph to fit the data in
x = top_25_keys
y = top_25_values
plt.figure(figsize=(14,10))
plt.bar(x, y)
plt.xlabel("Top 25 Most Used Words in MacBeth", fontsize=20)
plt.ylabel("Frequency of Each Word", fontsize=20)
plt.title("Top 25 Words Used in MacBeth & Their Respective Frequencies", fontsize=25)
plt.show()

   
# Pseudo-code outline
#1. split up the string into a list of words
#2. from there, make a dictionary using unique words from the list
#3. iterate through the list by adding up the frequency of each unique word in the list and storing it with 
#   the correct key in the dictionary
#4. sort the dictionary by highest values, limit to 25 highest values
#5. create a bar graph with y = frequency of times a word was used, x = the word within the 25 top words of macbeth
```
